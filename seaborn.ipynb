```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#seaborn has build datasets
 
tips = sns.load_dataset('tips') 
tips.head()
```

```python
#vuisualize the data 
## hue is used to group and color data based on the value of a category.
# sns.relplot()  a Seaborn function used for creating relational plots, like scatter plots and line plots, with support for advanced features like hue, col, row, etc.
#  Key Features:

# x, y: Data columns for axes

# hue: Color by category

# style: Shape of markers by category

# size: Size of markers by value

# kind: 'scatter' (default) or 'line'

# col/row: Create subplots by category

sns.relplot(data = tips ,  x = 'total_bill', y = 'tip', hue = 'smoker', size = 'size')
```
