```python
#matplotlib
#Use of making a plot
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
print(x)
print(y)
print(z)
```

```python
# Let's examine the arrays we created
print("X array (first 10 values):", x[:10])
print("X array shape:", x.shape)
print("Y (sin) array (first 10 values):", y[:10])
print("Z (cos) array (first 10 values):", z[:10])

# Create a plot to visualize the sine and cosine waves
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue')
plt.plot(x, z, label='cos(x)', color='red')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Sine and Cosine Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Python', 'Java', 'C++', 'JavaScript']
popularity = [85, 75, 65, 80]
ax.bar(languages, popularity, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')    
plt.title('Programming Languages Popularity')
plt.show()
```

```python
# pi chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
sizes = [15, 30, 45, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.show()
```
